```python
import cv2
import matplotlib.pyplot as plt
```

```python
def procesar_y_mostrar_imagen(ruta_imagen):
    # Cargar imagen en BGR
    img_bgr = cv2.imread(ruta_imagen)

    if img_bgr is None:
        print(f"No se pudo cargar la imagen: {ruta_imagen}")
        return

    # Convertir a RGB para visualización
    img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)

    # Convertir a HSV
    img_hsv = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2HSV)
    h, s, v = cv2.split(img_hsv)

    # Ecualizar el canal de valor
    v_eq = cv2.equalizeHist(v)
    img_hsv_eq = cv2.merge([h, s, v_eq])

    # Volver a BGR y luego a RGB
    img_eq_bgr = cv2.cvtColor(img_hsv_eq, cv2.COLOR_HSV2BGR)
    img_eq_rgb = cv2.cvtColor(img_eq_bgr, cv2.COLOR_BGR2RGB)

    # Mostrar ambas imágenes
    plt.figure(figsize=(10, 5))

    plt.subplot(1, 2, 1)
    plt.imshow(img_rgb)
    plt.title('Imagen original')
    plt.axis('off')

    plt.subplot(1, 2, 2)
    plt.imshow(img_eq_rgb)
    plt.title('Imagen mejorada (Ecualización V)')
    plt.axis('off')

    plt.tight_layout()
    plt.show()
```

```python
procesar_y_mostrar_imagen('/content/P5i.jpg')
procesar_y_mostrar_imagen('/content/p5i1.png')
procesar_y_mostrar_imagen('/content/p5i2.jpg')
```
